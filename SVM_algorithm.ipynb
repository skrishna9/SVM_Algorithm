```python
from sklearn.datasets import load_digits
```

```python
digits =load_digits()
```

```python
import pandas as pd 
```

```python
df =pd.DataFrame(digits.data,columns=digits.feature_names)
```

```python
df.head()
```

```python
df['target']=digits.target
```

```python
df.head()
```

```python
df[df.target==1].head()
```

```python
digits.target_names
```

```python
#now adding one more column name as pixel_names
df['pixel_names']=df.target.apply(lambda x:digits.target_names[x])
```

```python
df.head()
```

```python
from matplotlib import pyplot as plt
%matplotlib inline 
df0=df[df.target==0]
df1=df[df.target==1]
df2=df[df.target==2]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(['target','pixel_names'],axis='columns')
```

```python
x.head()
```

```python
y=df.target
```

```python
y
```

```python
#now tranning our model 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.svm import SVC
```

```python
model=SVC(kernel='linear')
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
